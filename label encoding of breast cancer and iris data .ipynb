```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv("data.csv") #we have loaded the data for breast cancer
```

```python
data.head()
```

```python
#let us see how many malignant and benign rows in this data
data["diagnosis"].value_counts()
```

```python
#there are 357 benign and 212 malignant in this dataset
```

```python
#now apply label encoder
label_encode= LabelEncoder()
```

```python
label=label_encode.fit_transform(data.diagnosis)#we are tranforming m and b to 0s and 1s
```

```python
label
```

```python
data["targets"]=label #we are appending one column named target and storing 0s and 1sto it
```

```python
data.head()
```

```python
data["targets"].value_counts()
```

```python
#ok u might be confused that 0 should be malignant and 1should benign...in label encoder data will arrange acc to alphabetical
#order here b comes first so 0 and m comes afetr b so 1 thats it
```

```python
#this is label encoding for breast cancer 
```
